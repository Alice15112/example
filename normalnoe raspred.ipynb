```python


import numpy as np
from matplotlib import pyplot as plt
import statistics

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
x = np.arange(-5, 5, 0.1)
mean = statistics.mean(x)
p = np.pi
ax.plot(x, 1 / ((2 * p) ** 0.5 / 2) * np.exp(-(x - mean)** 2 / 2 * 4), label='σ = 1/2')
ax.plot(x, 1 / ((2 * p) ** 0.5) * np.exp(-(x - mean)** 2 / 2), label='σ = 1')
ax.plot(x, 1 / ((2 * p) ** 0.5 * 2) * np.exp(-(x - mean)** 2 / 2 / 4), label='σ = 2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("Плотность нормального распределения")
ax.legend();
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = int(input())
m = int(input())
fig, axs = plt.subplots(figsize=(7, 5), nrows=n, ncols=m, layout="tight")
fig.suptitle("Роза")
for i in range(n):
    for j in range(m):
        f = np.linspace(-8*np.pi, 8*np.pi, 1000)
        r = np.sin(((j + 1) / (i + 1) * f))
        x = r * np.cos(f)
        y = r * np.sin(f)
        axs[i][j].plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import statistics
from mpl_toolkits import mplot3d
from matplotlib.colors import Normalize

x = np.linspace(-5, 5, 10)
y = x.copy().T
xg, yg = np.meshgrid(x, y)
z = np.sin((xg ** 2 + yg ** 2) ** 0.5)

fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot_surface(xg, yg, z, cmap='plasma', )
print(" ")
fig.colorbar(plt.cm.ScalarMappable(norm=Normalize(0, 1), cmap='plasma'),
             ax=ax)

fig1, axs = plt.subplots()
axs.contourf(xg, yg, z,  cmap='plasma')

fig1.colorbar(plt.cm.ScalarMappable(norm=Normalize(0, 1), cmap='plasma'),
             ax=axs)
```
